```python
from datetime import date

class Patient_Management_System:
    def __init__(self,Name,Age,Gender,DOB,Address,Hospital_Number):
        self.Name = Name
        self.Age = Age
        self.Gender = Gender
        self.DOB = DOB
        self.Address = Address
        self.Hospital_Number = Hospital_Number
        
    def Accept(self):
        Name = input('Please enter you full name: ')
        Age = int(input('Please enter your age in years: '))
        YOB = int(input('Please enter your Year of Birth: '))
        Gender = input('Are you Male or Female: ')
        DOB = input('Please enter your date of birth as DD/MM/YY: ')
        Address = input('Please enter you residential address: ')
        Hospital_Number = input('Please enter your hospital number: ')
        if date.today().year - YOB != Age:
            print('FAKE AGE!!! YOB not corresponding with AGE!!! NOT ACCEPTED')
        elif Age > 6:
            print('ABOVE PAEDIATRIC AGE!! NOT ACCEPTED')
        else:
            print('ACCEPTED')

    def Display(self):
        print(f'Patient\'s Name: {Name}')
        print(f'Patient\'s Age: {Age}')
        print(f'Patient\'s Gender: {Gender}')
        print(f'Patient\'s DOB: {DOB}')
        print(f'Patient\'s Current Address: {Address}')
        print(f'Patient\'s Hospital Number: {Hospital_Number}')


patient = Patient_Management_System('Name','Age','Gender','DOB','Address','Hospital_Number')


print(patient.Accept())
print(patient.Display())
```
