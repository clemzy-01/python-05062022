## 2nd Loop is the 'while' loop

#### it is used when you want to meet a certain condition.

```python
import math
count = 0
num = int(input('Please provide a number '))
while num != 0:
    print(math.sqrt(num))
    count = count + 1
    num = int(input('Please provide a number: '))
```

```python
#Objective: to create 100 random number between 45 to 70

import random as rn
numbers = []
counter = 1
while counter <= 100:
    num = rn.randint(45,70)
    numbers.append(num)
    counter +=1 #('counter +=1' is the same as 'counter = counter + 1')
print(numbers)
```

# WRITING A DICE GAME

```python
# USING THE FOR LOOP TO WRITE THE DICE GAME
# remember to import all modules needed first
score1 = 0
score2 = 0

for i in range(1,11):
    dice1 = rn.randint(1,6)
    dice2 = rn.randint(1,6)
    
    if dice1 > dice2:
        print(f'''Player 1 score: {dice1}
Player 2 score: {dice2}
Player 1 WINS''')
        
        score1 = score1 + 3
    elif dice2 > dice1:
        print(f'''Player 1 score: {dice1}
Player 2 score: {dice2}
Player 2 WINS''')
        score2 = score2 + 3
    else:
        print(f'''Player 1 score: {dice1}
Player 2 score: {dice2}
IT's A DRAW''')
        score1 = score1 + 1
        score2 = score2 + 1

if score1 > score2:
    print(f'Congratulations, player 1 is the overall winner with total score of {score1}')
elif score2 > score1:
    print(f'Congratulations, player 2 is the overall winner with total score of {score2}')
else:
    print('IT\'S A DRAW')
            
```

```python
# USING THE WHILE LOOP TO WRITE THE DICE GAME
# remember to import all modules needed first
score1 = 0
score2 = 0
counter = 1

while counter <= 10:
    dice1 = rn.randint(1,6)
    dice2 = rn.randint(1,6)
    
    if dice1 > dice2:
        print(f'''Player 1 score: {dice1}
Player 2 score: {dice2}
Player 1 WINS''')
        
        score1 = score1 + 3
    elif dice2 > dice1:
        print(f'''Player 1 score: {dice1}
Player 2 score: {dice2}
Player 2 WINS''')
        score2 = score2 + 3
    else:
        print(f'''Player 1 score: {dice1}
Player 2 score: {dice2}
IT's A DRAW''')
    
        score1 = score1 + 1
        score2 = score2 + 1
    counter +=1

if score1 > score2:
    print(f'Congratulations, player 1 is the overall winner with total score of {score1}')
elif score2 > score1:
    print(f'Congratulations, player 2 is the overall winner with total score of {score2}')
else:
    print('IT\'S A DRAW')
            
```
